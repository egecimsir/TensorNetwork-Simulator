```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
dft = lambda N: N**2
fft = lambda N: N * np.log2(N)
qft = lambda N: np.log2(N) ** 2
```

```python
x = np.linspace(1, 100, 10**3)
```

```python
plt.figure(figsize=(8, 6))

# plt.suptitle("Big-O Complexity", fontsize=16)
plt.ylim(top=250)
plt.grid()
plt.xlabel("Input Size (N)", fontsize=16)
plt.ylabel("Time Complexity", fontsize=16)

plt.plot(x, dft(x), label="DFT")
plt.plot(x, fft(x), label="FFT")
plt.plot(x, qft(x), label="QFT")

plt.legend()
plt.show()
```
